```python
import pandas as pd
import numpy as np

columns = [
    "duration","protocol_type","service","flag","src_bytes","dst_bytes","land",
    "wrong_fragment","urgent","hot","num_failed_logins","logged_in","num_compromised",
    "root_shell","su_attempted","num_root","num_file_creations","num_shells",
    "num_access_files","num_outbound_cmds","is_host_login","is_guest_login","count",
    "srv_count","serror_rate","srv_serror_rate","rerror_rate","srv_rerror_rate",
    "same_srv_rate","diff_srv_rate","srv_diff_host_rate","dst_host_count",
    "dst_host_srv_count","dst_host_same_srv_rate","dst_host_diff_srv_rate",
    "dst_host_same_src_port_rate","dst_host_srv_diff_host_rate","dst_host_serror_rate",
    "dst_host_srv_serror_rate","dst_host_rerror_rate","dst_host_srv_rerror_rate",
    "label","difficulty"
]

df = pd.read_csv("../data/KDDTrain+.txt", names=columns)
df["binary_label"] = df["label"].apply(lambda x: "normal" if x == "normal" else "attack")

df.shape
```

```python
df = df.drop(columns=["label", "difficulty"])
df.shape
```

```python
for col in ["duration", "src_bytes", "dst_bytes"]:
    df[col] = np.log1p(df[col])

df[["duration", "src_bytes", "dst_bytes"]].describe()
```

```python
df = pd.get_dummies(df, columns=["protocol_type", "service", "flag"], drop_first=True)
df.shape
```

```python
df["binary_label"] = df["binary_label"].map({"normal": 0, "attack": 1})
df["binary_label"].value_counts()
```

```python
df.info()
```

```python
df.to_csv("../data/processed_train.csv", index=False)
```

## Summary

- Dropped `label` (redundant with binary_label) and `difficulty` (metadata, not a real feature).
- Applied log1p transformation to duration, src_bytes, dst_bytes to fix right-skew.
- One-hot encoded protocol_type, service, and flag (42 -> 120 columns).
- Encoded binary_label: normal=0, attack=1.
- Saved cleaned dataset to data/processed_train.csv for use in Day 5.
